### Inverted condition filter
~ : Tilde (~) is a Python negation operator which reverses the condition. For example, ~ON becomes OFF.

Now, if we want to see the developer whose salary of less than 40000, instead of writing the condition again, we just put a ~ operator on a high_salary_filt (~ high_salary_filt). Isn't that awesome….

```python
df.loc[~ high_salary_filter, ['Country', 'ConvertedComp', 'LanguageWorkedWith']].head(20)
```

### Complex filter using multiple conditions
This section will use Logical (AND, OR) Operators. Before jumping to the complex filtering, I want to introduce you to logical operators AND and OR operators.

AND: It returns TRUE only if all conditions are true.
OR: It returns TRUE only if either one operand is true.
In Pandas, & is equal to AND and | is equivalent to OR
In our example, the filter is defined as follows:

Salary of greater than 50000 in ConvertedComp

AND

The Country column should contain value to the United States

AND

LanguageWorkedWith column string values contain ‘Python.’ — (we will use str. contains() method for this condition. This is a new method we are learning here. It is used to test if a pattern or sub string is contained within a string value.)

All these conditions will be applied to the loc method as a filter, and the rows that meet the requirements should return columns ‘Country’, ‘ConvertedComp’, ‘LanguageWorkedWith’.

```python
filt = ((df['ConvertedComp'] > 50000) & (df['Country'] == 'United States')) | (df['LanguageWorkedWith'].str.contains('Python'))
df.loc[filt, ['Country', 'ConvertedComp', 'LanguageWorkedWith']].head(50)
```
