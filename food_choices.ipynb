Sobre las elecciones alimentarias de estudiantes universitarios; el objetivo fue intentar comprender cuán importante es la información nutricional.

```python
import pandas as pd
import numpy as np
```

```python
import re

def limpiar_peso(valor):
    valor = str(valor)
    
    solo_numeros = re.sub(r"\D", "", valor)
    if(len(solo_numeros)==0):
        return np.nan
    else:
        return int(solo_numeros)
    
def limpiar_gpa(valor):
    valor = str(valor)
    
    # Eliminamos cualquier cosa que NO sea un dígito [0-9] o un punto [.]
    solo_numeros = re.sub(r'[^0-9.]', "", valor)
    
    if len(solo_numeros) == 0:
        return np.nan
    else:
        # Usamos float() para mantener los decimales del GPA
        return float(solo_numeros)
```

```python
df = pd.read_csv('food_coded.csv')
```

```python
df['weight'] = df['weight'].apply(limpiar_peso)
df['weight'].unique()
```

```python
df['weight'] = df['weight'].fillna(df['weight'].median())
df['weight'].unique()
```

# RECONOCER OUTLIERS WEIGHT

```python
df['weight'].describe()
```

```python
q1 = df['weight'].describe()['25%']
q3 = df['weight'].describe()['75%']
iqr = q3-q1

lower_fence = q1 - (1.5*iqr)
upper_fence = q3 + (1.5*iqr)
print(f"LOWER: {lower_fence}, UPPER: {upper_fence}")
```

```python
condicion = (df['weight'] < 67.5) | (df['weight'] > 247.5)
df = df[~ condicion]
df
```
