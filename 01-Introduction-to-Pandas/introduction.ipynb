### Series
A Series is a one-dimensional array with a single column that can hold heterogeneous data (integers, strings, floating-point numbers, Python objects, etc.).It has an index that can be used to call different data values.

Series function in its basic form only needs a list of data to be created.

pd.Series(data)

```python
data = [2, 3, 4, 5]
series_1 = pd.Series(data)
series_1
```

```python
series_1 = pd.Series(data = data, name = 'Hours')
series_1
```

```python
sale = pd.Series([500,85000,40000,9600,100000], index = ['Cookie', 'Choco', 'Toothpaste', 'Soap', 'Toffee'])
sale
```

### Data Frame
Data frame is a two-dimensional array. It can also be thought of as a two or more Series. It looks like a spreadsheet or tabular data with rows, columns and an index. It has two axes, axis 0 and 1. 0 always means rows, and 1 means columns.

The most straightforward way to create a data frame is with keys and corresponding lists from the dictionary. Index by default is numbers (0,1,2..n) unless index defines explicitly.

Let's create a dictionary first.

```python
data = {'country': ['Canada', 'Canada', 'US', 'US', 'UK', 'UK'],
'year': [2010, 2011, 2012, 2013, 2014, 2015],
'pop': [2.5, 1.7, 2.6, 1.4, 4.9, 3.7]}
data
```

```python
dataf = pd.DataFrame(data)

dataf
```

Reorder columns

```python
pd.DataFrame(data, columns=['year', 'country', 'pop'])
```

Add a column which doesn't exists

```python
data = pd.DataFrame(data, columns= ['year', 'country', 'pop', 'uni'])
data
```
