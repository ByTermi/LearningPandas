### View all columns in a data set
#### Set Option method

As observed in the above output, it does not show all the columns. The default (20 columns) behavior shows a truncated view of the columns and rows. Often we want to get a glance at all the columns and their data. To achieve this, we can use the pd.set_option() method with its option — display.max_columns.

```python
pd.set_option('display.max_columns', 61)

df
```

### Quick view of the first and the last few rows
#### Head and tail methods

The head() method returns the first five while the tail() function returns the last five rows of the data set by default. If you like to see more or fewer rows, you can specify any number between brackets ().

```python
df.head(3)
```

```python
df.tail(3)
```

shape() is a data frame attribute that returns the size of the data set with the total number of rows and the number of columns (rows, columns) in the data set.

```python
df.shape
```

### Info method

The info() method shows a concise summary of the data frame as the name indicates. It returns the data frame Index, the total number of rows & columns, columns names, column data types, non-null values and count of different data types.

Non-null values are essential to note here. They show much data is in each column is use able. Null values are not that useful, and you might want to replace them with some data or remove those rows altogether. We will review data cleaning techniques in a future article.

```python
df.info()
```

### Descriptive statistics summary
#### Describe method

The describe() method is used to see the summary of the descriptive statistics. It summarizes a data set and shows the output with central tendency (mean), dispersion (standard deviation), percentile (25th, 50th, 75th), count, min and max.

```python
df.describe()
```

### Random data sampling
#### Sample method

The sample() method returns a random number of rows and columns from the data frame. By default, it returns one row, but we can input any number in the method brackets.

```python
df.sample()
```

```python
df.sample(3)
```

### Unique values and frequency tables
#### Unique method

The unique() method returns a unique value of a column. Null values are excluded by default.

```python
df['Hobbyist'].unique()
```

### Value counts method

value_counts() method takes things a step further. It shows unique values in a column as well as their count. The output is a concise and useful frequency table in descending order.

```python
df['Hobbyist'].value_counts()
```
